```python
def tweets_for_hashtags(consumer_key, consumer_secret, access_token, access_token_secret, hashtags):
    
    #create authentication for accessing Twitter
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_token_secret)

    #initialize Tweepy API
    api = tweepy.API(auth)
    
    
    tweets = []
    for hashtag_phrase in hashtags:
        #for each tweet matching our hashtags, write relevant info to the spreadsheet
        for tweet in tweepy.Cursor(api.search, q=hashtag_phrase+' -filter:retweets', \
                                   lang="en", tweet_mode='extended').items(100):
            #print(tweet.full_text,tweet.created_at)
            tweets.append(tweet.,tweet.full_text)
        
    return tweets
           # w.writerow([tweet.created_at, tweet.full_text.replace('\n',' ').encode('utf-8'), tweet.user.screen_name.encode('utf-8'), [e['text'] for e in tweet._json['entities']['hashtags']], tweet.user.followers_count])
```

```python
def get_freq_words(word_count,size):
    words = sorted(word_count.items(), key = 
             lambda kv:(kv[1], kv[0]), reverse=True)
    #print(words)
    freq_words = []
    count = 0
    new_word_count = {}
    for index in range(len(words)):
        if(len(words[index][0]) != 1 and str.isalnum(words[index][0]) and words[index][0] not in stopwords):
            count += 1
            freq_words.append(words[index][0])
            new_word_count[words[index][0]] = words[index][1]
        if(count == size):
            break
    return freq_words,new_word_count
        
```

```python

'''This function returns the freq words occuring in the entire dataset i.e after combining all the tweets, and also returns the sparse vectors for each tweet which tell which of the words in the frequent words occur in which tweet(indicated by a "1") '''
def get_BOW_vector(tweets,hashtags):
    word_count = {}
    for tweet in tweets:
        words = nltk.word_tokenize(tweet)
        for word in words:
            if word in hashtags:
                words.remove(word)
            
            if(word not in word_count.keys()):
                word_count[word] = 1
            else:
                word_count[word] += 1
    freq_words,new_word_count = get_freq_words(word_count,len(word_count))
    
    
    vector_array = []
    for tweet in tweets:
        vector = []
        for word in freq_words:
            if word in nltk.word_tokenize(tweet):
                vector.append(1)
            else:
                vector.append(0)
        vector_array.append(vector)
    return vector_array,freq_words,new_word_count

vector_array, freq_words ,freq_word_count = get_BOW_vector(tweets,hashtags[0:2])
```
