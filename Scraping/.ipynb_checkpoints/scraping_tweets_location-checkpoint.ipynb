```python
def get_location(username):
    try:
        userOBJ =api.get_user(username)
        res= userOBJ.location
    except:
        res= "user suspended"
    return res
```

```python

locations = list(withLoc_df['location'])
tweets = list(withLoc_df['Tweet_text'])


def getCountryTweets():
    global locations,tweets,country_tweets
    countries = [str(word).lower() for word in list(cities['country'])]
    subcountries = [str(word).lower() for word in list(cities['subcountry'])]
    citiess = [str(word).lower() for word in list(cities['name'])]
    abb1 = [str(word).lower() for word in list(cities['abb1'])]
    abb2 = [str(word).lower() for word in list(cities['abb2'])]
    
    
    for i in range(len(locations)):
        
        location = locations[i]
        words = location.split(', ')
        word = words[-1].lower()
        flag = 0
        if word == "bangalore":
            word = "bengaluru"
        if word in countries or word in abb1 or word in abb2:
            

            if word not in countries:
                if word in abb1:
                    country = countries[abb1.index(word)]
                else:
                    country = countries[abb2.index(word)]
            else:
                country = word
#                 locations = list(withLoc_df['location'])

            tweet = tweets[i]
            country_tweets[country].append(tweet)
            #print(country,tweet)
            flag = 0
            
        elif word in subcountries:
            country = countries[subcountries.index(word)]
            
            tweet = tweets[i]
            country_tweets[country].append(tweet)
            #print(country,tweet)
            flag = 0
            

        elif word in citiess:
            country = countries[citiess.index(word)]
            
            tweet = tweets[i]
            country_tweets[country].append(tweet)
            #print(country,tweet)
            flag = 0
             
        else:
            flag = 1
        
        if flag:
            word = words[0].lower()

            if word == "bangalore":
                word = "bengaluru"
            if word in countries or word in abb1 or word in abb2:
              
                if word not in countries:
                    if word in abb1:
                        country = countries[abb1.index(word)]
                    else:
                        country = countries[abb2.index(word)]
                else:
                    country = word
#                 locations = list(withLoc_df['location'])

                tweet = tweets[i]
                country_tweets[country].append(tweet)
                #print(country,tweet)
                flag = 0
                
            elif word in subcountries:
                country = countries[subcountries.index(word)]
                
                tweet = tweets[i]
                country_tweets[country].append(tweet)
                #print(country,tweet)
                flag = 0
                

            elif word in citiess:
                country = countries[citiess.index(word)]
                
                tweet = tweets[i]
                country_tweets[country].append(tweet)
                #print(country,tweet)
                flag = 0
                  
            else:
                flag = 1
        
        if flag:
            words = word_tokenize(location)
            flag = 0

            for w in words:
                word = w.lower()
                if word == "bangalore":
                    word = "bengaluru"
                if word in countries or word in abb1 or word in abb2:
                    
                    if word not in countries:
                        if word in abb1:
                            country = countries[abb1.index(word)]
                        else:
                            country = countries[abb2.index(word)]
                    else:
                        country = word
    #                 locations = list(withLoc_df['location'])

                    tweet = tweets[i]
#                     print(country,tweet)
                    #country_tweets[country].append(tweet)

                    flag = 0
                    break
                elif word in subcountries:
                    country = countries[subcountries.index(word)]
                    
                    tweet = tweets[i]
                    country_tweets[country].append(tweet)
                    #print(country,tweet)
                    flag = 0
                    break

                elif word in citiess:
                    country = countries[citiess.index(word)]
                    
                    tweet = tweets[i]
                    country_tweets[country].append(tweet)
                    #print(country,tweet)
                    flag = 0
                    break  
                else:
                    flag = 1

            if flag:
                country = location + "(not in csv)"
                if country not in country_tweets:
                    country_tweets[country] = list()
                tweet = tweets[i]
                country_tweets[country].append(tweet)
                #print(country,tweet)


                

getCountryTweets()
    
```
