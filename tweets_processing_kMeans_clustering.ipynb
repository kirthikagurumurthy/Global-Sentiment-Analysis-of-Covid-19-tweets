```python
positive_words_list = set(opinion_lexicon.positive())
negative_words_list = set(opinion_lexicon.negative())
tokenizer = treebank.TreebankWordTokenizer()
def getSentimentValue(review):
    senti = 0
    words = [lemmatizer.lemmatize(word.lower()) for word in tokenizer.tokenize(review)]
    for word in words:
#         print(word)
        if word in positive_words_list and word in final_BOW:
            senti += 1
        elif word in negative_words_list and word in final_BOW:
            senti -= 1
    return senti

def Distance(x1,x2):
    return (x1-x2)**2
```

```python

def kmeans(tweets,sent_scores):
    count = 0
#     print(p1,p2,"HEYYYY")
    while(True):
        p1 = random.randrange(len(tweets))
        p2 = random.randrange(len(tweets))
        if(p1 != p2):
            break
#     print(p1,p2,"HEYYYY")
    
    
    point1 = sent_scores[p1]
    point2 = sent_scores[p2]
    
#     print("points",point1,point2,sent_scores[p1],sent_scores[p2])
    
    cluster1 = {p1:point1}

    cluster2 = {p2:point2}

    old_centroid1 = new_centroid1 = sent_scores[p1]

    old_centroid2 = new_centroid2 = sent_scores[p2]



    while(True):
        print("iteration ",count)
        
#         print("initial centroids: ",old_centroid1,old_centroid2)
        old_centroid1 = new_centroid1

        old_centroid2 = new_centroid2

        for i in range(len(tweets)):

            point = sent_scores[i]

            if(Distance(old_centroid1,point) > Distance(old_centroid2,point)):
                if(tweets[i] not in cluster2):
                    cluster2[tweets[i]] = point
                    if tweets[i] in cluster1:
                        cluster1.pop(tweets[i])
            else:
                if(tweets[i] not in cluster1):
                    cluster1[tweets[i]] = point
                    if tweets[i] in cluster2:
                        cluster2.pop(tweets[i])
            
            if(len(cluster1) > 0):
                new_centroid1 = sum(cluster1.values())/len(cluster1)
#                 print(new_centroid1,old_centroid1)
            else:
                
                new_centroid1 = 0
            
            if(len(cluster2) > 0):
                new_centroid2 = sum(cluster2.values())/len(cluster2)
#                 print(new_centroid2,old_centroid2)
            else:
                
                new_centroid2 = 0
        
        count += 1
        
        if(old_centroid1 == new_centroid1 and old_centroid2 == new_centroid2):
            break
            
    print("Final centroids: ",new_centroid1,new_centroid2,"\n")
    return cluster1,cluster2,new_centroid1,new_centroid2
    
    
```

```python
def labelClusters(cluster1,cluster2,new_centroid1,new_centroid2):
    if(new_centroid1 < new_centroid2):
        return cluster2,cluster1

    else:
        return cluster1,cluster2
```

```python

for country,tweets in country_tweets.items():
    sent_scores = []
    
    print("for country",country,":\n")
    
    for tweet in tweets:
#             print("hi",tweet)
        sent_scores.append(getSentimentValue(tweet))
#             print("hey",tweet)
#         print("hi")
    if len(tweets) == 1:
        if sent_scores[0] >= 0:
            pos_tweets = tweets
            neg_tweets = []
        else:
            pos_tweets = []
            neg_tweets = tweets
    else:
        cluster1,cluster2,new_centroid1,new_centroid2 = kmeans(tweets,sent_scores)
        pos_cluster,neg_cluster = labelClusters(cluster1,cluster2,new_centroid1,new_centroid2)
        pos_tweets = list(pos_cluster.keys())
        neg_tweets = list(neg_cluster.keys())
        
    if(len(neg_tweets) > len(pos_tweets)):
        for i in range(len(pos_tweets),len(neg_tweets)):
            pos_tweets.append("")
    else:
        for i in range(len(neg_tweets),len(pos_tweets)):
            neg_tweets.append("")

    df = pd.DataFrame({"Positive":pos_tweets,"Negative":neg_tweets})
    
    if "(not in csv)" not in country:
        df.to_csv("Classified_Country_tweets/"+country+".csv")
    else:
        name = country.split('/')
        name = ' '.join(name)
        df.to_csv("Classified_UnknownCountry_tweets/"+name+".csv")
    
```
