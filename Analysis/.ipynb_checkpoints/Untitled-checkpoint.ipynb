```python
import nltk
from nltk.tokenize import treebank
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer 
import math
from sklearn.model_selection import train_test_split
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import random
from nltk.corpus import opinion_lexicon
from nltk.corpus import wordnet as wn
from nltk.stem import WordNetLemmatizer 
```

```python
positive = pd.read_csv("../Datasets/Classified_Country_tweets/Positive.csv")
neutral = pd.read_csv("../Datasets/Classified_Country_tweets/Neutral.csv")
negative = pd.read_csv("../Datasets/Classified_Country_tweets/Negative.csv")
```

```python
frames = [positive,neutral,negative]
all_frames = pd.concat(frames)
```

```python
all_frames
```

```python
def vaderSentiment(value):
    if value >= 0.2:
        return("positive")
    
    elif value < -0.2:
        return("negative") 
    
    else : 
        return("Neutral") 
```

```python
def vaderSentimentValue(review):
    sid_obj =  SentimentIntensityAnalyzer()
    
    sentiment_dict = sid_obj.polarity_scores(str(review))
    
    return sentiment_dict['compound']
```

```python
all_frames['Vader-sentiment value'] = all_frames['Tweets'].apply(vaderSentimentValue)
all_frames['Vader-sentiment'] = all_frames['Vader-sentiment value'].apply(vaderSentiment)
```
