```python
import pandas as pd
import os
```

```python
april_may = pd.read_csv('../Datasets/TweetsWithLocation/corona_april_may_withLoc.csv',sep='\t')
```

```python
jan_feb1 = pd.read_csv('../Datasets/TweetsWithLocation/tweets_with_loc_JanFeb.csv')
```

```python
jan_feb2 = pd.read_csv('../Datasets/TweetsWithLocation/tweets_with_loc_JanFebTS.csv')
```

```python
june = pd.read_csv('../Datasets/TweetsWithLocation/tweets_with_loc_June.csv')
```

```python
march = pd.read_csv('../Datasets/TweetsWithLocation/TweetsWithLoc_Mar.csv')
```

```python
june['Tweet_text'] = june['tweet_txt']
```

```python
march['Tweet_text'] = march['text']
```

```python
jan_feb2['Tweet_text'] = jan_feb2['text']
jan_feb2['date'] = jan_feb2['timestamp']
```

```python
june['date'] = june['date-time']
```

```python
march['date'] = march['timestamp']
```

```python
frames = [jan_feb1,jan_feb2,march,april_may,june]
tweets_date = {}
```

```python
for frame in frames:
    tweets = list(frame['Tweet_text'])
    dates = list(frame['date'])
    for i in range(len(tweets)):
        if tweets[i] not in tweets_date:
            tweets_date[tweets[i]] = dates[i]
    
```

```python
tweets_date
```

```python

directory = r'../Datasets/Classified_Country_tweets'
for filename in os.listdir(directory):
    
    df = pd.read_csv(directory+"/"+filename)
    pos_tweets = list(df['Positive'])
    if(str(pos_tweets[0]).replace('.', '', 1).isdigit()):
        pos_tweets.pop(0)
   
    pos_dates = []
    for tweet in pos_tweets:
        if(tweet in tweets_date):
            pos_dates.append(str(tweets_date[tweet]))

    pos_tweets = pos_tweets[0:len(pos_dates)]
    
    neg_tweets = list(df['Negative'])
    if(str(neg_tweets[0]).replace('.', '', 1).isdigit()):
        neg_tweets.pop(0)
    neg_dates = []
    for tweet in neg_tweets:
        if(tweet in tweets_date):
            neg_dates.append(str(tweets_date[tweet]))
    neg_tweets = neg_tweets[0:len(neg_dates)]
            
    neu_tweets = list(df['Neutral'])
    if(str(neu_tweets[0]).replace('.', '', 1).isdigit()):
        neu_tweets.pop(0)
    neu_dates = []
    for tweet in neu_tweets:
        if(tweet in tweets_date):
            neu_dates.append(str(tweets_date[tweet]))
    neu_tweets = neu_tweets[0:len(neu_dates)]   
            
    pos_df = pd.DataFrame({"Tweets":pos_tweets,"Date":pos_dates})
    neg_df = pd.DataFrame({"Tweets":neg_tweets,"Date":neg_dates})
    neu_df = pd.DataFrame({"Tweets":neu_tweets,"Date":neu_dates})
    country = filename[0:-4]
    pos_df.to_csv("../Datasets/Classified_Country_Tweets/Positive/"+country+".csv")
    neg_df.to_csv("../Datasets/Classified_Country_Tweets/Negative/"+country+".csv")
    neu_df.to_csv("../Datasets/Classified_Country_Tweets/Neutral/"+country+".csv")
    
```
