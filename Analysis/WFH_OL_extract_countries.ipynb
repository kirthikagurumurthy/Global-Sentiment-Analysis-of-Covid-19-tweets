```python
import numpy
import pandas as pd
import re
```

```python
wfh = pd.read_csv("../Datasets/WFH_all.csv")
```

```python
ol = pd.read_csv("../Datasets/OL.csv")
```

```python
cities = pd.read_csv("../Datasets/cities.csv")
```

```python
city_country = {}
```

```python
countries = [string.lower() for string in list(cities['country'])]
subcountries = [string.lower() for string in list(cities['subcountry'])]
city = [string.lower() for string in list(cities['name'])]
abb1 = [str(string).lower() for string in list(cities['abb1'])]
abb2 = [str(string).lower() for string in list(cities['abb2'])]
```

```python
for i in range(len(subcountries)):
    if(subcountries[i].lower() not in city_country):
        city_country[subcountries[i].lower()] = countries[i].lower()
    if(city[i] not in city_country):
        city_country[city[i]] = countries[i].lower()
    if(abb1[i] not in city_country):
        city_country[abb1[i]] = countries[i].lower()
    if(abb2[i] not in city_country):
        city_country[abb2[i]] = countries[i].lower()
    city_country[countries[i]] = countries[i]
```

```python
locations = [str(string).lower() for string in list(wfh['location'])]
```

```python
country_column = []
```

```python
len(locations)
```

```python
city
```

```python
for location in locations:
    flag = 0
    if "nan" == location:
        flag = 1
        country_column.append("NA")
    else:   
        split1 = re.split(', | ',location)
        for word in split1:
            if word in city_country:
                country_column.append(city_country[word])
                flag = 1
                break
    if(not flag):
        split2 = re.split(", ",location)
        for word in split2:
            if word in city_country:
                country_column.append(city_country[word])
                flag = 1
                break
    if(not flag):
        split3 = re.findall("[A-Za-z]*[ ]*[a-z]*",location)
        for word in split3:
            if word in city_country:
                country_column.append(city_country[word])
                flag = 1
                break
                
    if(not flag):
        country_column.append("NA")
            
```

```python
len(country_column)-country_column.count("NA")
```

```python
len(country_column)
```

```python
wfh["country"] = country_column
```

```python
len(wfh)
```

```python
wfh
```

```python
wfh.to_csv('../Datasets/WFH_all.csv')
```
