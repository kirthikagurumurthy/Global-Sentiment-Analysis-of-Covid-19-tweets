```python
country_tweets = {}
directory = r'Classified_Country_tweets'
pos_count = {}
neg_count = {}
for filename in os.listdir(directory):
    df = pd.read_csv(directory+"/"+filename)
    pos_list = [str(tweet) for tweet in list(df['Positive']) if(str(tweet) != 'nan')]
    neg_list = [str(tweet) for tweet in list(df['Negative']) if(str(tweet) != 'nan')]
    
    if(len(pos_list) > 0 and len(neg_list) > 0):
        if(pos_list[0].isnumeric() and neg_list[0].isnumeric()):
            pos_count[filename[0:-4]] = len(pos_list)-1
            neg_count[filename[0:-4]] = len(neg_list)-1
        else:
            pos_count[filename[0:-4]] = len(pos_list)
            neg_count[filename[0:-4]] = len(neg_list)
    elif(len(pos_list) > 0):
#     print(type(neg_list[0]))
        if(pos_list[0].isnumeric()):
            pos_count[filename[0:-4]] = len(pos_list)-1
            neg_count[filename[0:-4]] = len(neg_list)
        else:
            pos_count[filename[0:-4]] = len(pos_list)
            neg_count[filename[0:-4]] = len(neg_list)
    elif(len(neg_list) > 0):
        if(neg_list[0].isnumeric()):
            pos_count[filename[0:-4]] = len(pos_list)
            neg_count[filename[0:-4]] = len(neg_list)-1
        else:
            pos_count[filename[0:-4]] = len(pos_list)
            neg_count[filename[0:-4]] = len(neg_list)
    else:
        pass
    
    country_tweets['positive'] = pos_count
    country_tweets['negative'] = neg_count
```
