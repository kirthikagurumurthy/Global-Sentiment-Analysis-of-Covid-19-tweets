```python
def plot_map(df, col, pal):
    df = df[df[col]>0]
    fig = px.choropleth(df, locations="Country/Region", locationmode='country names', 
                  color=col, hover_name="Country/Region", 
                  title=col, hover_data=[col], color_continuous_scale=pal)
#    fig.update_layout(coloraxis_showscale=False)
    fig.show()
```

```python
df=pd.read_csv('final_dataset.csv')
df
```

```python
df=df.drop(['Unnamed: 0'],axis=1)
```

```python
df['country'] = df['country'].replace(['united states'], 'us')
```

```python
df['Sentiment'] = df['Sentiment'].str.lower()
```

```python
df_country=pd.DataFrame(df.groupby(['country']).size())
df_country=df_country.reset_index()
df_country
```

```python
for i in df_country['country']:
    df1=df[df['country']==i]
    neg=len(df1[df1.Sentiment=='negative'])
    pos=len(df1[df1.Sentiment=='positive'])
    neu=len(df1[df1.Sentiment=='neutral'])
    df_country.loc[df_country['country']==i,'negative']=neg
    df_country.loc[df_country['country']==i,'positive']=pos
    df_country.loc[df_country['country']==i,'neutral']=neu
df_country
```

```python
df_country['negative_n']=df_country['negative']/df_country[0]
df_country['positive_n']=df_country['positive']/df_country[0]
df_country['neutral_n']=df_country['neutral']/df_country[0]
df_country
```

```python
df_country.to_csv('../Datasets/final_countrywise.csv')
```

```python
country_wise['country_lower'] = country_wise['Country/Region'].str.lower()
df_merged = country_wise.merge(df_country, left_on="country_lower", right_on="country", how="left")
df_merged
```

```python
df.to_csv('final.csv')
df_country.to_csv('groupedbycountry.csv')
```

```python
df_merged.rename(columns = {'positive_n':'normalised positive tweets'}, inplace = True)
df_merged.rename(columns = {'negative_n':'normalised negative tweets'}, inplace = True)
df_merged.rename(columns = {'neutral_n':'normalised neutral tweets'}, inplace = True)
plot_map(df_merged, 'normalised positive tweets', 'matter')
```

```python
plot_map(df_merged, 'normalised negative tweets', 'matter')
```

```python
plot_map(df_merged, 'normalised neutral tweets', 'matter')
```

```python
pos_df=df[df.Sentiment=='positive']
neg_df=df[df.Sentiment=='negative']
neu_df=df[df.Sentiment=='neutral']
```

```python
count_series = pos_df.groupby(['Country/Region', 'Date']).size()
newpos_df = count_series.to_frame(name = 'size').reset_index()
newpos_df=newpos_df.sort_values(by='Date')
newpos_df=newpos_df.reset_index()
newpos_df
```

```python
# Positive WFH tweets Over the time

fig = px.choropleth(newpos_df, locations="Country/Region", 
                    color=np.log(newpos_df["size"]),
                    locationmode='country names', hover_name="Country/Region", 
                    animation_frame=newpos_df["Date"].dt.strftime('%Y-%m-%d'),
                    title='Positive Sentiment over time', color_continuous_scale=px.colors.sequential.matter)
fig.update(layout_coloraxis_showscale=False)
fig.show()
```

```python
count_series = neg_df.groupby(['Country/Region', 'Date']).size()
newneg_df = count_series.to_frame(name = 'size').reset_index()
newneg_df=newneg_df.sort_values(by='Date')
newneg_df=newneg_df.reset_index()
newneg_df
```

```python
# Negative tweets Over the time

fig = px.choropleth(newneg_df, locations="Country/Region", 
                    color=np.log(newneg_df["size"]),
                    locationmode='country names', hover_name="Country/Region", 
                    animation_frame=newneg_df["Date"].dt.strftime('%Y-%m-%d'),
                    title='Negative Sentiment over time', color_continuous_scale=px.colors.sequential.matter)
fig.update(layout_coloraxis_showscale=False)
fig.show()
```

```python
count_series = neu_df.groupby(['Country/Region', 'Date']).size()
newneu_df = count_series.to_frame(name = 'size').reset_index()
newneu_df=newneu_df.sort_values(by='Date')
newneu_df=newneu_df.reset_index()
# Neutral tweets Over the time

fig = px.choropleth(newneu_df, locations="Country/Region", 
                    color=np.log(newneu_df["size"]),
                    locationmode='country names', hover_name="Country/Region", 
                    animation_frame=newneu_df["Date"].dt.strftime('%Y-%m-%d'),
                    title='Neutral Sentiment over time', color_continuous_scale=px.colors.sequential.matter)
fig.update(layout_coloraxis_showscale=False)
fig.show()
```
