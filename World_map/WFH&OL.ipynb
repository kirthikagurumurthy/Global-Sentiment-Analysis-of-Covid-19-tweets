```python
def plot_map(df, col, pal):
    df = df[df[col]>0]
    fig = px.choropleth(df, locations="Country/Region", locationmode='country names', 
                  color=col, hover_name="Country/Region", 
                  title=col, hover_data=[col], color_continuous_scale=pal)
#    fig.update_layout(coloraxis_showscale=False)
    fig.show()
```

```python
df_wfh=pd.read_csv('../Datasets/WFH_all.csv')
df_ol=pd.read_csv('../Datasets/OL.csv')
df_wfh
```

```python
df_wfh=df_wfh.dropna()
df_ol=df_ol.dropna()
```

```python
df_ol=df_ol.reset_index(drop=True)
df_wfh=df_wfh.reset_index(drop=True)
```

```python
#df_ol=df_ol.drop(['Unnamed: 0','Unnamed: 0.1','Unnamed: 0.1.1','Unnamed: 0.1.1.1','Unnamed: 0.1.1.1.1'], axis=1)
df_wfh=df_wfh.drop(['Unnamed: 0','Unnamed: 0.1','Unnamed: 0.1.1','Unnamed: 0.1.1.1','Unnamed: 0.1.1.1.1'], axis=1)
```

```python
df_ol=df_ol.drop(['Unnamed: 0.1.1.1','index'], axis=1)
df_ol
```

```python
#df_wfh=df_wfh.drop(['index'], axis=1)
df_wfh
```

```python
df_ol['country'] = df_ol['country'].replace(['united states'], 'us')
df_wfh['country'] = df_wfh['country'].replace(['united states'], 'us')
```

```python
df_ol_country=pd.DataFrame(df_ol.groupby(['country']).size())
df_ol_country=df_ol_country.reset_index()
df_ol_country
```

```python
for i in df_ol_country['country']:
    df=df_ol[df_ol['country']==i]
    neg=len(df[df.Sentiment=='Negative'])
    pos=len(df[df.Sentiment=='Positive'])
    neu=len(df[df.Sentiment=='Neutral'])
    df_ol_country.loc[df_ol_country['country']==i,'negative']=neg
    df_ol_country.loc[df_ol_country['country']==i,'positive']=pos
    df_ol_country.loc[df_ol_country['country']==i,'neutral']=neu
df_ol_country
```

```python
country_wise['country_lower'] = country_wise['Country/Region'].str.lower()
df_merged_ol = country_wise.merge(df_ol_country, left_on="country_lower", right_on="country", how="left")
df_merged_ol
```

```python
df_merged_ol.rename(columns = {'positive':'positive tweets OL'}, inplace = True)
df_merged_ol.rename(columns = {'negative':'negative tweets OL'}, inplace = True)
df_merged_ol.rename(columns = {'neutral':'neutral tweets OL'}, inplace = True)
plot_map(df_merged_ol, 'positive tweets OL', 'matter')
```

```python
df_wfh['country'] = df_wfh['country'].replace(['united states of america'], 'us')
df_ol['country'] = df_ol['country'].replace(['united states of america'], 'us')

df_wfh['Country/Region'] = df_wfh['country'].map(country_wise.set_index('country_lower')['Country/Region'])
df_ol['Country/Region'] = df_ol['country'].map(country_wise.set_index('country_lower')['Country/Region'])
#df_wfh.timestamp=pd.DatetimeIndex(df_wfh.timestamp).tz_localize('UTC')
#df_ol.timestamp=pd.DatetimeIndex(df_ol.timestamp).tz_localize('UTC')
df_wfh['Date'] = df_wfh['timestamp'].astype(str).str[:10]
df_ol['Date'] = df_ol['timestamp'].astype(str).str[:10]
df_wfh['Date']=pd.to_datetime(df_wfh['Date']) 
df_ol['Date']=pd.to_datetime(df_ol['Date']) 
```

```python
pos_wfh_df=df_wfh[df_wfh.Sentiment=='Positive']
neg_wfh_df=df_wfh[df_wfh.Sentiment=='Negative']
neu_wfh_df=df_wfh[df_wfh.Sentiment=='Neutral']
pos_ol_df=df_ol[df_ol.Sentiment=='Positive']
neg_ol_df=df_ol[df_ol.Sentiment=='Negative']
neu_ol_df=df_ol[df_ol.Sentiment=='Neutral']
```

```python
count_series = pos_wfh_df.groupby(['Country/Region', 'Date']).size()
newposwfh_df = count_series.to_frame(name = 'size').reset_index()
newposwfh_df=newposwfh_df.sort_values(by='Date')
newposwfh_df=newposwfh_df.reset_index()
newposwfh_df
```

```python
# Positive WFH tweets Over the time

fig = px.choropleth(newposwfh_df, locations="Country/Region", 
                    color=np.log(newposwfh_df["size"]),
                    locationmode='country names', hover_name="Country/Region", 
                    animation_frame=newposwfh_df["Date"].dt.strftime('%Y-%m-%d'),
                    title='Positive WFH Sentiment over time', color_continuous_scale=px.colors.sequential.matter)
fig.update(layout_coloraxis_showscale=False)
fig.show()
```
